```python
import PIL
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFilter

import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random

def RandomPaste(origin, img):
    # Resize the pasted image
    width, height = origin.size
    propotion = random.uniform(0.1, 0.5)
    img_width = int(width * propotion)
    propotion = img_width/img.size[0]
    img_height = int(img.size[1] * propotion)
    if img_height > height:
        img_height = height
    img = img.resize((img_width, img_height), Image.ANTIALIAS)
    
    # Rotate the pasted image
    rotate_angle = random.randint(0, 360)
    img = img.rotate(rotate_angle, expand = True)
    img_width, img_height = img.size
    
    # Paste image
    r,g,b,a = img.split()
    img_x = int(random.uniform(0, 1) * (width - img_width))
    img_y = int(random.uniform(0, 1) * (height - img_height))
    origin.paste(img, (img_x, img_y), a)
    plt.figure(figsize=(20,40))
    plt.imshow(origin)
    
    # Add noise
    
    # Addjust Color Temperature
    
    # Addjust Color difference
    
    # Add Deviation
    heightOffset = random.randint(-20, 20)
    widthOffset = random.randint(-20, 20)
    
    
    return [img_x, img_y, img_width, img_height]
```

```python
background = Image.open(os.path.join("./img/background", random.choice(os.listdir("./img/background"))))
obj = []
for i in range(random.randint(1, 2)):
    obj.append(Image.open(os.path.join("./img/object", random.choice(os.listdir("./img/object")))))
```

```python
for i in obj:
    RandomPaste(background, i)
    
# plt.figure(figsize=(20,40))
# plt.imshow(background)
```
