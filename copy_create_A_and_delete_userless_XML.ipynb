```python
from shutil import copyfile
import os
import random
import glob as gb


def highwayPaste():
    img_path = "./highway/input"
    dst_path = "./highway/images"
    for i in range(130):
        img_name = random.choice(os.listdir(img_path))
        src = os.path.join(img_path, img_name)
        dst = os.path.join(dst_path, 'car_'+str(i)+'.jpg')
        copyfile(src, dst)
    
```
